```python
df3 = df2.drop(columns='canonical_smiles')
```

```python
smiles = []

for i in df2.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
df_clean_smiles = pd.concat([df3,smiles], axis=1)
df_clean_smiles
```

```python
!pip install rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df_clean_smiles.canonical_smiles)
df_lipinski
```

```python
df_combined = pd.concat([df2,df_lipinski], axis=1)
```

```python
df_combined
```

```python
import numpy as np

def stand_norm(input):
    stand_norm = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        stand_norm.append(-np.log10(molar))

    input['stand_norm'] = stand_norm
    x = input.drop('standard_value_norm', 1)
        
    return x
```

```python
df_combined.standard_value.describe()
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)
        
    return x
```

```python
df_norm = norm_value(df_combined)
df_norm
```

```python
df_norm.standard_value_norm.describe()
```

```python
df_final = stand_norm(df_norm)
df_final
```

```python
import pandas as pd
dataset = pd.read_csv('final_data.csv')
```

```python
dataset
```

```python
X = dataset.drop('class', axis=1)
x = dataset.drop('class', axis=1)
```

```python
X
```

```python
Y = dataset['class']
Y
```

```python
X.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.9 * (1 - .9)))    
X = selection.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import joblib
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=101)
```

```python
SVC_model = SVC( C=1.0, kernel='poly', degree=5, gamma='scale', coef0=0.0, shrinking=True, probability=True, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)
```

```python
SVC_model.fit(X_train,y_train)
```

```python
SVC_train_pre=SVC_model.predict(X_train)
```

```python
accuracy_score(y_train,SVC_train_pre)
```

```python
filename = 'SVC.sav'
joblib.dump(SVC_model, filename)
```

```python
SVC_test_predict = SVC_model.predict(X_test)
```

```python
print(classification_report(y_test,SVC_test_predict))
```

```python
target=['Cancer','Non-cancer']
report1 = classification_report(y_test,SVC_test_predict, target_names=target, digits=4,  output_dict=True)
df=pd.DataFrame(report1)
plt.rcParams["figure.figsize"] = (15,8)
df.iloc[:3, :9].T.plot(kind='bar')
plt.show()
```

```python
svc_cm=confusion_matrix(y_test,SVC_test_predict)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(svc_cm,annot=True,fmt='.0f',linewidths=3,square=True,cmap='Blues')
plt.xlabel('Actual label')
plt.ylabel('Predicted Label')
plt.title('Support Vector Machine',size=15)
print ("Support Vector Machine",  accuracy_score(y_test,SVC_test_predict))
```
